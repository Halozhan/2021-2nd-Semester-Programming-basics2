```python
data = pd.read_csv("sample.csv")
print(data)

#필터를 이용한 데이터 선택 : isin()
data1 = data["빅데이터분석"].isin([60,93])
print(data1) # 필터에 해당하는 값이 있으면 True, 그렇지 않으면 False 반환

# 실제 값을 반환하려면 DataFrame 내부에서 선택의 값으로 입력
data2 = data[data["빅데이터분석"].isin([60,93])]
data2

data3 = data[data.isin([60,98])] # 전체 데이터 중에서 60, 90 점을 가진 사람 위치찾기
data3
```

```python
# drop()으로 행 또는 열 삭제하기   axis = 0 : 행 삭제 ( default )   axis = 1 : 열 삭제
data = pd.read_csv("sample.csv")
print(data)

data1 = data.drop(2)
print(data1)

data2 = data.drop([5,7])
print(data2)

data3 = data.drop("빅데이터분석", axis= 1)
print(data3)
```

```python
#미리 만들어 둔 함수 적용
data = pd.read_csv('sample.csv')

def func(row):
    if row > 90:
        return "A"
    elif row > 80:
        return "B"
    else:
        return "C"
    
data["학점"] = data["프로그래밍기초"].apply(func)
data
```

```python
# insert()로 원하는 위치에 열 입력하기
data = pd.read_csv("sample.csv")

column_name = "새과목"
column_value = [57,54,53,80,84,88,79,64,75,65,84,90,72,95,100,94,85,63,57,95]
data.insert(1, column_name, column_value)
print(data)
```

```python
# append()로 행 추가하기
data = pd.read_csv("sample.csv")

append_data = pd.Series([100,200], ["빅데이터분석","SW공학"])
data1 = data.append(append_data, ignore_index=True)
print(data1)
```
