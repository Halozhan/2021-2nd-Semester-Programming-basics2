```python
import pandas as pd
data = pd.read_csv('sample.csv')
print(data)
data_1 = data.loc[0:3, ['빅데이터분석']] # 0 ~ 3
data_2 = data.iloc[0:3, [2]] # 0 ~ 2
data_3 = data.at[0, '빅데이터분석']
data_4 = data.iat[2, 1]
print(data_1)
print(data_2)
```

```python
data = pd.read_csv("sample.csv")
print(data)

data1 = data[data["빅데이터분석"]>=80]
print(data1)

data2 = data[(data["빅데이터분석"]>=60) & (data["프로그래밍기초"]>=80)]
print(data2)
```

```python
data = pd.read_csv("sample.csv")
print(data)

#필터를 이용한 데이터 선택 : isin()
data1 = data["빅데이터분석"].isin([60,93])
print(data1) # 필터에 해당하는 값이 있으면 True, 그렇지 않으면 False 반환

# 실제 값을 반환하려면 DataFrame 내부에서 선택의 값으로 입력
data2 = data[data["빅데이터분석"].isin([60,93])]
data2

data3 = data[data.isin([60,98])] # 전체 데이터 중에서 60, 90 점을 가진 사람 위치찾기
data3
```

```python
#열 값을 변경하려면 열 이름을 선택하고 같은 길이의 데이터를 리스트나 Series로 입력
data = pd.read_csv("sample.csv")
print(data)

data["AI활용"] = [57,54,53,80,84,88,79,64,75,65,84,90,72,95,100,94,85,63,57,95]
print(data)
```

```python
#열 목록에 없는 이름 입력하면 새로운 데이터가 추가된다
data = pd.read_csv("sample.csv")
print(data)

data["데이터베이스"] = [57,54,53,80,84,88,79,64,75,65,84,90,72,95,100,94,85,63,57,95]
print(data)
```

```python
data = pd.read_csv("sample.csv")
print(data)

print(data.loc[0, "프로그래밍기초"])
data.loc[0, "프로그래밍기초"] = 100
print(data.loc[0, "프로그래밍기초"])
```
