```python
data = pd.read_csv("sample.csv")

plt.figure(num = "성적그래프")
plt.rcParams.update({'font.family': 'malgun gothic', 'font.size': 15})

plt.title("학년별 성적")

# plt 적용, 한개의 화면에서 나타나도록 하기 위한 axes 설정
# ax에 plt 데이터
ax = plt.gca()

xcolumns = ["학번", "학번", "학번"]
ycolumns = ["프로그래밍기초", "빅데이터분석", "AI활용"]

color = ['#209FDF','#99CA53','#F6A625', '#6D5FD5','#BF593E', '#FF0000', '#0000FF']

data.plot(kind = "scatter", x = "학번", y = "프로그래밍기초", ax = ax, color = color[0], s = 10, marker = "P")
data.plot(kind = "scatter", x = "학번", y = "빅데이터분석", ax = ax, color = color[2], s = 30, marker = "D")
data.plot(kind = "scatter", x = "학번", y = "AI활용", ax = ax, color = color[4], s = 50, marker = "*")

plt.xlabel("학번")
plt.ylabel("성적")

plt.show()
```
