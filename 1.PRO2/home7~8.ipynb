```python
# 과제 7
import pandas as pd

data = pd.read_csv("sample.csv")

def scoring(row):
    if row >= 90:
        return "A"
    elif row >= 80:
        return "B"
    else:
        return "C"

# 1
# 총합
total = data.프로그래밍기초 + data.빅데이터분석 + data.AI활용

data["총합"] = total
# 평균
data["평균"] = round(total / 3, 2)

# 2
data["순위"] = data.rank(axis="index", method="min", numeric_only=True, ascending=False)["총합"]

# 3
data = data.sort_values(axis=0, by="순위")

# 4
data["학점"] = data["평균"].apply(scoring)

# 출력
data
```

```python
# 과제 8
# 1
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams.update({'font.family': 'malgun gothic','font.size': 12}) # font

plt.title("학생별 성적")

plt.xlabel("학번")
plt.ylabel("성적")

ax = plt.gca()

data.plot(kind="bar", x="학번", y="평균", ax=ax)
plt.xticks(rotation = 45)

plt.show()


# 2
plt.title("성적별 빈도수")

plt.xlabel("성적")
plt.ylabel(None)

ax = plt.gca()

plt.hist(data["학점"], bins = 5)

plt.show()
```
